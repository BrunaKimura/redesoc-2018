Análise do banco de dados de Cabreúva

```python
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib
```

```python
cabreuva = pd.read_table("cabreuva_limpa.csv", encoding="windows-1252", sep=",")
```

```python
cabreuva
```

```python
cabreuva.dtypes
```

```python
cabreuva.EVANGELICO = cabreuva.EVANGELICO.astype('category')
```

```python
cabreuva.EVANGELICO.cat.categories = ('não', 'Sim')
```

```python
cabreuva.BEMCONSIGOMESMO.value_counts()
```

```python
cabreuva[["EVANGELICO", "BEMCONSIGOMESMO"]][1:5]
```

```python
ct = pd.crosstab(cabreuva.EVANGELICO, cabreuva.BEMCONSIGOMESMO)
ct
```

A relação entre relegião e se sentir bem consigo mesmo não apresentou nenhuma relação, já que os valores estão bem próximos.

```python
cabreuva.FILHALÉSBICA = cabreuva.FILHALÉSBICA.astype('category')
cabreuva.FILHALÉSBICA.cat.categories = ('discordo', 'discordo parcialmente', 'concordo parcialmente', 'concordo')
```

```python
ct = pd.crosstab(cabreuva.EVANGELICO, cabreuva.FILHALÉSBICA, normalize=True, margins=True)
ct
```

```python
ct2 = pd.crosstab(cabreuva.FILHALÉSBICA, cabreuva.EVANGELICO)
```

```python
ct2 = (ct2/ct2.sum().sum()*100).round(decimals=0)
ct2.plot(kind='bar', color=('yellow', 'salmon'))
plt.ylabel('Frequência Relativa (em %)')
```

   Neste gráfico de barras, fica bastante claro a relação entre a religião e a aceitação de uma filha lésbica. Apesar da grande maioria estar entre discordo em partes e concordo em partes. Pode-se ver que as pessoas que discordam ou disconcordam em partes é maioria evangélicos. Já os que concordam e corcondam em partes não são evangélicos.
